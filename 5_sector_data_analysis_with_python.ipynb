```python
# ================= FINANCE SECTOR =================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

salary=[40,72,40,120,70,56,54,48,81]
experience=[5,6,5,10,7,4,3,2,9]

print(f"Mean: {np.mean(salary)}")
print(f"Median: {np.median(salary)}")
print(f"Mode: {pd.Series(salary).mode()[0]}")
print(f"Variance: {np.var(salary)}")
print(f"Standard Deviation: {np.std(salary)}")
print(f"Correlation: {np.corrcoef(salary,experience)}")

data=pd.DataFrame(
    {
        "salary":salary,
        "experience":experience
    }
)

print(data)

plt.figure(figsize=(10,5))
sns.scatterplot(data=data,x="experience",y="salary")
plt.title("Salary vs Experience")
plt.show()

plt.figure(figsize=(10,5))
sns.lineplot(data=data,x="experience",y="salary")
plt.title("Salary vs Experience")
plt.show()

plt.figure(figsize=(10,5))
sns.barplot(data=data,x="experience",y="salary")
plt.title("Salary vs Experience")
plt.show()

plt.figure(figsize=(10,5))
sns.histplot(data=data,x="salary",kde=True)
plt.title("Salary Distribution")
plt.show()

plt.figure(figsize=(10,5))
plt.pie(data["salary"],labels=data["experience"],autopct="%1.1f%%")
plt.title("Salary Pie Chart")
plt.show()
```

```python
# ================= RETAIL SECTOR =================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sales=[20,35,40,55,70,60,45,30,50]
customers=[100,150,180,220,300,260,200,140,240]

print(f"Mean: {np.mean(sales)}")
print(f"Median: {np.median(sales)}")
print(f"Mode: {pd.Series(sales).mode()[0]}")
print(f"Variance: {np.var(sales)}")
print(f"Standard Deviation: {np.std(sales)}")
print(f"Correlation: {np.corrcoef(sales,customers)}")

data=pd.DataFrame(
    {
        "sales":sales,
        "customers":customers
    }
)

print(data)

plt.figure(figsize=(10,5))
sns.scatterplot(data=data,x="customers",y="sales")
plt.title("Sales vs Customers")
plt.show()

plt.figure(figsize=(10,5))
sns.lineplot(data=data,x="customers",y="sales")
plt.title("Sales vs Customers")
plt.show()

plt.figure(figsize=(10,5))
sns.barplot(data=data,x="customers",y="sales")
plt.title("Sales vs Customers")
plt.show()

plt.figure(figsize=(10,5))
sns.histplot(data=data,x="sales",kde=True)
plt.title("Sales Distribution")
plt.show()

plt.figure(figsize=(10,5))
plt.pie(data["sales"],labels=data["customers"],autopct="%1.1f%%")
plt.title("Sales Pie Chart")
plt.show()
```
