```python
import statistics
import matplotlib.pyplot as plt
members_names=["harsha","Riya","Krishna","Shiva","Esha"]
ages=[21,25,24,23,21]
workout_hours=[2,1.5,2,1,3]
calories_burned=[700,350,600,300,750]
print("---Fitness dataset---")
for i in range(len(members_names)):
  print(members_names[i],ages[i],workout_hours[i],calories_burned[i])

#mean workout Hours
avg_hours=sum(workout_hours)/len(workout_hours)
print(f"\nAverage Workout hours: {avg_hours}")

#maximum Calories
print(f"Maximum calories burned: {max(calories_burned)}")

#minimum Calories
print(f"Minimum calories burned: {min(calories_burned)}")

#median Calories
print(f"Median calories burned: {statistics.median(calories_burned)}")

#standard deviation
print(f"Standard deviaion : {statistics.stdev(calories_burned)}")

#most active member
index=calories_burned.index(max(calories_burned))
print(f"Most active member: {members_names[index]}")

plt.bar(members_names,calories_burned)
plt.xlabel("Members")
plt.ylabel("Calories Burned")
plt.title("Fitness analysis")
plt.show()
```
