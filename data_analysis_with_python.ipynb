```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

marks=[70,80,90,60,100]
hours=[2,3,5,1,6]

print(F"Mean: {np.mean(marks)}")
print(F"Median: {np.median(marks)}")
print(F"Mode: {pd.Series(marks).mode()[0]}")
print(f"Variance: {np.var(marks)}")
print(f"Standard Deviation: {np.std(marks)}")
print(f"Correlation: {np.corrcoef(marks,hours)}")

data=pd.DataFrame(
    {
        "marks":marks,
        "hours":hours
    }
)

print(data)

plt.figure(figsize=(10,5))
sns.scatterplot(data=data,x="hours",y="marks")
plt.title("Marks vs Hours")
plt.show()

plt.figure(figsize=(10,5))
sns.lineplot(data=data,x="hours",y="marks")
plt.title("Marks vs Hours")
plt.show()

plt.figure(figsize=(10,5))
sns.barplot(data=data,x="hours",y="marks")
plt.title("Marks vs Hours")
plt.show()

plt.figure(figsize=(10,5))
sns.histplot(data=data,x="marks",kde=True)
plt.title("Marks Distribution")
plt.show()

plt.figure(figsize=(10,5))
plt.pie(data["marks"],labels=data["hours"],autopct="%1.1f%%")
plt.title("Pie Chart ")
plt.show()

```
