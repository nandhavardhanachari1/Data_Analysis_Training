```python
import pandas as pd

data=pd.DataFrame({
    "name":["vishnu","sunil","sivaram","nandha","vardhan"],
    "age":[40,50,78,68,36],
    "oxygen":[78,98,45,98,1],
    "condition":["Critical", "Stable", "Severe", "Moderate", "Critical"],


})
beds=2
```

```python
def calculate_priority(row):
    p = 0
    if row['age'] > 60:
        p = p + 10
    p = p + (row['age'] / 100)
    if row['condition'] == 'Critical':
        p = p + 5
    elif row['condition'] == 'Severe':
        p = p + 4
    elif row['condition'] == 'Moderate':
        p = p + 3
    elif row['condition'] == 'Mild':
        p = p + 2
    elif row['condition'] == 'Stable':
        p = p + 1
    p = p + (100 - row['oxygen']) / 10

    return p

data['priority'] = data.apply(calculate_priority, axis=1)

sorted_data = data.sort_values(by='priority', ascending=False)

print("Patients by priority:")
print(sorted_data)

final_bed_allocation = ['No'] * len(sorted_data)
for i in range(beds):
    if i < len(sorted_data):
        final_bed_allocation[i] = 'Yes'
sorted_data['final_bed_allocation'] = final_bed_allocation

print("Bed Allocation Result:")
print(sorted_data)
```
