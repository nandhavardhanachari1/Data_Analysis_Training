```python
import pandas as pd
df=pd.read_excel("/content/e_com_filled.xlsx")
print(df)
```

```python
print(df.info())
```

```python
#Diagnostic Analysis Project
customer_names = df['Customer_Name'].unique()
print(customer_names)
internet_failures=df[df['Internet']=='No']
print(internet_failures)
payment_failures=df[df['Payment Failure']=='Yes']
print(payment_failures)
delivery_charges=df[df['Delivery Charges']>50]
print(delivery_charges)
purchase_completed=df[df['Order_Status']=='Completed']
print(purchase_completed)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.countplot(x='Order_Status', data=df)
plt.title('Distribution of Order Status')
plt.xlabel('Order Status')
plt.ylabel('Count')
plt.show()
```

```python

df['Order_Status'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title('Proportion of Order Statuses')
plt.ylabel('')
plt.show()
```

```python

sns.histplot(df['Order_Price'], bins=5, kde=True)
plt.title('Distribution of Order Price')
plt.xlabel('Order Price')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python

sns.scatterplot(x='Order_Price', y='Delivery Charges', data=df)
plt.title('Order Price vs Delivery Charges by Order Status')
plt.xlabel('Order Price')
plt.ylabel('Delivery Charges')
plt.tight_layout()
plt.show()
```

```python

df_sorted = df.sort_values(by='Order_Date')
sns.lineplot(x='Order_Date', y='Order_Price', data=df_sorted, marker='o', color='purple')
plt.title('Order Price Trend Over Time')
plt.xlabel('Order Date')
plt.ylabel('Order Price')
plt.show()
```
